```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Generate synthetic time series data (random walk)
np.random.seed(42)
n=100
data=np.cumsum(np.random.randn(n)) # Cumulative sum to stimulate a trend

# Plot time series data
plt.figure(figsize=(10,4))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sm.graphics.tsa.plot_acf(data, lags=20, ax=axes[0])
axes[0].set_title("Autocorrelation Fuction (ACF)")

sm.graphics.tsa.plot_pacf(data, lags=20, ax=axes[1])
axes[1].set_title("Partial Autocorrelation Function(PACF)")
plt.show()
```
