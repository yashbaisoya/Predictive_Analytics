```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate time series data
np.random.seed(42)
time = np.arange(1, 101)

# Trend Component (Linear Growth)
trend = 0.5 * time

# Seasonality Component (Sine Wave)
seasonality = 5 * np.sin(2 * np.pi * time / 12)

# Cyclic Component (Long-term Fluctuation)
cycle = 10 * np.sin(2 * np.pi * time / 50)

# Irregular Component (Random Noise)
irregular = np.random.normal(scale=3, size=len(time))

# Total Time Series Data (Combination of Components)
time_series = trend + seasonality + cycle + irregular

# Plot the components
plt.figure(figsize=(10, 8))

plt.subplot(5, 1, 1)
plt.plot(time, trend, label="Trend", color='r')
plt.title("Trend Component")
plt.legend()

plt.subplot(5, 1, 2)
plt.plot(time, seasonality, label="Seasonality", color='g')
plt.title("Seasonality Component")
plt.legend()

plt.subplot(5, 1, 3)
plt.plot(time, cycle, label="Cyclic Variation", color='b')
plt.title("Cyclic Component")
plt.legend()

plt.subplot(5, 1, 4)
plt.plot(time, irregular, label="Irregular Component", color='purple')
plt.title("Irregular Component (Random Noise)")
plt.legend()

plt.subplot(5, 1, 5)
plt.plot(time, time_series, label="Combined Time Series", color='black')
plt.title("Full Time Series with All Components")
plt.legend()

plt.tight_layout()
plt.show()
```
